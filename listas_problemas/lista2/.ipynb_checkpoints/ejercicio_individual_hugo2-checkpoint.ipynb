##Apartado A

En este apartado separaremos los datos de entrenamiento de los de test del dataset recién importado.
Esta partición se hace haciendo uso del parametro stratify el cual preserva la proporcion de target similar al dataset original en los datasets de entrenamiento y test.
Esto lo hacemos para obtener mejores predicciones y reproductibilidad de los experimentos llevados acabo


```python
from sklearn.model_selection import train_test_split
#Separamos nuesra variable objetivo del resto y hacemos la separación 60/40
#Esta separación se hace estratificada para que se preserven las proporciones de target


X = appendicitis_data.drop("target",axis=1)
y = appendicitis_data.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.40, random_state=105, stratify=y)

print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)

X_train.head()
```


##Apartado B

Ahora partiendo de esta partición de datos hemos de realizar distintos preprocesos para hacer funcionar tres tipos de modelos diferentes:

-Naïve Bayes

-Regresión Logística

-K nearest neighbours

Para todos estos hemos de entrenar un modelo a partir de los datos preprocesados y obtener los siguentes indicadores:

-Validación cruzada

-Acierto del test

-Matriz de confusion

-Informe de clasificacion

Para este apartado y los siguentes haremos uso de una serie de funciones que realizarán operaciones genéricas para no repetir demasiado el código para cada modelo.

Estas haran directamente el print de los aspectos observados y formarán las confussion matrix para ser printadas en un futuro.


```python
from sklearn.metrics import  ConfusionMatrixDisplay,\
                  classification_report,  RocCurveDisplay, PrecisionRecallDisplay,\
                    accuracy_score, f1_score, precision_score, recall_score, confusion_matrix

from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV,RandomizedSearchCV,GridSearchCV

from sklearn.discriminant_analysis import LinearDiscriminantAnalysis, QuadraticDiscriminantAnalysis
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier

from yellowbrick.target.feature_correlation import feature_correlation
from yellowbrick.classifier import precision_recall_curve
from sklearn.preprocessing import StandardScaler,KBinsDiscretizer



def generate(X_train_sd, X_test_sd, y_train, y_test, model_name, model, ax):
    # entrenamos el modelo y generamos la predicción
    s_model = model.fit(X_train_sd, y_train)
    y_pred = s_model.predict(X_test_sd)

    # generamos cada uno de los indicadores a obtener

    # Validacion cruzada
    cross_val = np.mean(cross_val_score(model, X_train_sd, y_train, cv=5))
    print('The cross validation error average for ', model_name, ' is: ', cross_val.round(3))

    # Acierto del test
    test_acc = s_model.score(X_test_sd, y_test)
    print('The test error of ', model_name, ' is: ', test_acc.round(3))

    # Informe de clasificación
    class_inf = classification_report(y_pred, y_test, target_names=['0', '1'])
    print('The classification report of ', model_name, ' is:')
    print(class_inf)

    # Matrices de confusion
    ConfusionMatrixDisplay.from_estimator(s_model, X_test_sd, y_test, display_labels=['0', '1'], ax=ax)
    ax.set_title(f'Confusion Matrix - {model_name}')

```

Comenzamos preprocesando los datos simplemente escalando las variables para adecuar las mismas a los modelos que vamos a utilizar. Esto lo hacemos mediante copias para mantener siempre una copia base de la partición del dataset.

```python
from sklearn.preprocessing import StandardScaler
#preprocesamos escalando los datos
X_train_sd = X_train.copy(deep=True)
X_test_sd = X_test.copy(deep=True)
sc = StandardScaler()
X_train_sd = sc.fit_transform(X_train_sd)
X_test_sd = sc.transform(X_test_sd)
```

A continuación exploramos los hiperparametros más convenientes para nuestros modelos y generamos los resultados para comenzar a evaluarlos.

Los modelos los vamos aplicando a nuestra función genérica en un bucle para poder expandir de cara a futuras exploraciones el problema.

```python
# buscamos los hiperparametros para knn
knn = KNeighborsClassifier()
param = {'n_neighbors': np.arange(1, 20)}
knn_gs = GridSearchCV(knn, param, cv=5)
knn_gs.fit(X_train_sd, y_train)
print("los parametros encontrados por Grid Search:", knn_gs.best_params_['n_neighbors'])

models = [
    ("naive_bayes", GaussianNB()),
    ("logistic_regression", LogisticRegression()),
    ("KNN", KNeighborsClassifier(n_neighbors=knn_gs.best_params_['n_neighbors']))
]

fig, ax = plt.subplots(nrows=1, ncols=len(models), figsize=(5 * len(models), 5))

for i, (model_name, model) in enumerate(models):
    generate(X_train_sd, X_test_sd, y_train, y_test, model_name, model, ax[i])


plt.tight_layout()
plt.show()
```

Empecemos el estudio de como se comportan los modelos sin discretizar para después comparar estos a los valores discretizados.

En lo que a todos los modelos respecta los resultados son moderadamente satisfactorios con errores de validación curzada superiores a 80% y errores en el cojunto de pruebas superiores al 80% en todos los casos e incluso al 90% en el caso de la Regresión Logica.

Dependiendo del problema elegiriamos uno u otro pues todos tienen resultados similares. Todo dependa de que métricas queramos monitorizar en nuestro problema. Básicamente en este caso considero que un falso negativo es muchísimo más grave que un falso positivo (al tratarse de un diagnostico es más crítico pasar por alto una enfermedad y matar un paciente que asustar a alguién para luego dar una buena noticia). En este caso todos tienen un fallo bastante mayor al tratarse de los positivos cosa que hace difícil justificar usarlos para un caso real sin un entrenamiento mucho mas exhaustivo.

####Informe de clasificación

En general a todas las clases les cuesta más predecir los positivos que los negativos (se equivocan más en la clase 1 que en la clase 0). La base con la que trabajamos es pequeña así que puede ser que le falte contenido para predecir mejor (y el numero de enfermos es mucho menor que el de gente sana por lo tanto un pequeño cambio puede tener grandes implicaciones porcentuales). La que mejores puntajes tiene de todo en general es la Regresión Logistica. A pesar de esto el error de validación cruzada es menor que el del conjunto de pruebas cosa que es preocupante pero también puede ser achacada al insuficiente tamaño del conjunto de test.

####Matrices de confusion
En las matrices de confusión se ve mejor todo esto. Al ser el conjunto de positivos tan pequeño uno o dos falsos negativos tienen un alto aporte al porcentaje de fallos mientras que de falsos positivos hay mas pero porcentualmente (y a nivel conceptual teniendo en cuenta el problema también son menos importantes). El que se equivoca menos en general es el Linear Regression.


##Apartado C

Ahora en este apartado aplicaremos métodos de discretización haciendo uso de el preproceso KBinsDiscretizer de Scikit-learn. Para ello mirando de aproximar la distribución de datos originales.

Este K bins básicamente se encarga de discretizar mediante creación de k elementos de clasificación, cómo recipientes ("bins" en inglés) de intervalos.

Estos pueden ser codificados de diferentes maneras (one hot, ordinales etc.). En partícular a nostoros nos piden Onehot-dense para que la codificación se haga con un one-hot encoding y se devuelva en un array denso. De esa forma cada valor se corresponde a un atributo el conjunto de datos.

Estos se puede hacer con uso de dos métodos:

-Quantile: La discretización se hace para los valores de los quantiles, lo que significa que cada recipiente tenga aproximadamente el mismo número de elementos

-Kmeans: La distcretización se hace basandose en los centroides del proceso de clustering KMeans

Estos tienen diferentes fuertes según como los datos se distribuyan.

Haremos esto mismo para 2, 3 y 4 recipientes.

Ajustaremos y compararemos los resultados respecto al apartado anterior

Por cada combinación generaremos hiperparametros de KNN y entonces obtendremos los resultados para ser comentados más tarde.

En este apartado miro de intentar optimizar al máximo el código para no escribir una cantidad exagerada del mismo y sea díficil de leer y trabajar con el. Creo funciones nuevas como generate_brief para limitar el output y lo genero todo a la vez.

```python
def generate_brief(X_train_sd, X_test_sd, y_train, y_test, model_name, model):
    # entrenamos el modelo y generamos la predicción
    s_model = model.fit(X_train_sd, y_train)
    y_pred = s_model.predict(X_test_sd)

    # generamos solo la validación cruzada y el acierto porque hay muchísimos ejemplos a comparar

    # Validacion cruzada
    cross_val = np.mean(cross_val_score(model, X_train_sd, y_train, cv=5))

    # Acierto del test
    test_acc = s_model.score(X_test_sd, y_test)

    ret = {
            "cross_validation_error_avg":cross_val,
            "test_error":test_acc,
        }

    return ret


strats = ['quantile','kmeans']
bins = [2,3,4]
results = {}

for strat in strats:
    aux_results = {}
    for n_bin in bins:

        #Discretizamos acorde al caso
        X_train_bin = X_train.copy(deep=True)
        X_test_bin = X_test.copy(deep=True)
        kBinDiscretizer = KBinsDiscretizer(n_bins=n_bin, encode='onehot-dense', strategy=strat)
        kBinDiscretizer.fit(X_train)
        X_train_bin = kBinDiscretizer.transform(X_train_bin)
        kBinDiscretizer.fit(X_test)
        X_test_bin = kBinDiscretizer.transform(X_test_bin)

        # buscamos los hiperparametros para knn
        knn = KNeighborsClassifier()
        param = {'n_neighbors': np.arange(1, 20)}
        knn_gs = GridSearchCV(knn, param, cv=5)
        knn_gs.fit(X_train_bin, y_train)
        print("los parametros encontrados por Grid Search:", knn_gs.best_params_['n_neighbors'])

        #para cada modelo dentro de cada subcaso generamos los resultados y los guardamos
        models = [
            ("naive_bayes", GaussianNB()),
            ("logistic_regression", LogisticRegression()),
            ("KNN", KNeighborsClassifier(n_neighbors=knn_gs.best_params_['n_neighbors']))
        ]

        aux_results2 = {}

        for i, (model_name, model) in enumerate(models):
            aux_results2[model_name] = generate_brief(X_train_bin, X_test_bin, y_train, y_test, model_name, model)

        aux_results[n_bin] = aux_results2

    results[strat] = aux_results
```

```python
# Convert the dictionary to a DataFrame
results_list = []
for discretization_method, values in results.items():
    for num_bins, models in values.items():
        for model, metrics in models.items():
            results_list.append([discretization_method, num_bins, model, metrics['cross_validation_error_avg'], metrics['test_error']])

# Create the DataFrame
columns = ['Discretization Method', 'Num Bins', 'Model', 'Cross Validation Error Avg', 'Test Error']
df = pd.DataFrame(results_list, columns=columns)

# Sort the DataFrame by 'Test Error' column in descending order
sorted_df = df.sort_values(by='Test Error', ascending=False)

sorted_df
```

Viendo los resultados es interesante pues podemos observar que realmente si cambia radicalmente la calidad de los modelos (ahora exploraremos si para bien o para mal).

Como anteriormente los que mejor se comportan con el cocnjunto de test son las regresiones logísticas con KNN siguiendo de muy cerca (las regresiones logísticas siguen teniendo el problema de tener un error de validación cruzada peor al de test y se debe a los mismos motivos que antes).

La diferencia entre estos resultados y los originales son en algunos casos una ligera mejoría en la validación cruzada o un empeoramiento generalizado.

En algunos como naïve bayes funciona terriblemente cuando estamos haciendo uso de kmeans obteniendo unos resultados mucho peores en general y para su mejor configuración (con quantile y un número bajo de cestos) obtiene un resultado inferior al obtenido de forma simple.

En general ninguno mejora mucho sobre el tipo de base. Deberiamos explorar más en profundidad (más bins tal vez o diferentes hiperparametros) pero no parece que esta sea una alternativa que suponga un gran cambio de paradigma.

```python
# Boxplot de errores de test por modelo
plt.figure(figsize=(10, 6))
sns.boxplot(x='Model', y='Test Error', data=sorted_df, palette='Set3')
plt.title('Boxplot of Test Errors by Model')
plt.xlabel('Model')
plt.ylabel('Test Error')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

En estos boxplots se puede ver mejor la tendencia de los resultados en el caso del error de test. Aquí vemos a un claro perdedor siendo naive_bayes. Esto tiene sentido pues es el más simple de los procedimientos.

```python
fig, axs = plt.subplots(1, 2, figsize=(14, 6))

# Test Error vs. Number of Bins para 'kmeans'
kmeans_df = sorted_df[sorted_df['Discretization Method'] == 'kmeans']
kmeans_df_sorted = kmeans_df.sort_values('Num Bins')

for model in kmeans_df_sorted['Model'].unique():
    model_data = kmeans_df_sorted[kmeans_df_sorted['Model'] == model]
    axs[0].plot(model_data['Num Bins'], model_data['Test Error'], marker='o', label=model)

axs[0].set_title('Test Error vs. Number of Bins (KMeans Discretization)')
axs[0].set_xlabel('Number of Bins')
axs[0].set_ylabel('Test Error')
axs[0].legend()
axs[0].grid(True)

quantile_df = sorted_df[sorted_df['Discretization Method'] == 'quantile']
quantile_df_sorted = quantile_df.sort_values('Num Bins')

for model in quantile_df_sorted['Model'].unique():
    model_data = quantile_df_sorted[quantile_df_sorted['Model'] == model]
    axs[1].plot(model_data['Num Bins'], model_data['Test Error'], marker='o', label=model)

axs[1].set_title('Test Error vs. Number of Bins (Quantile Discretization)')
axs[1].set_xlabel('Number of Bins')
axs[1].set_ylabel('Test Error')
axs[1].legend()
axs[1].grid(True)

plt.tight_layout()
plt.show()
```

Observemos un poco en este gráfico como afecta esta discretizacion a los modelos a medida que azanvamos en el numero de bins que utilizamos para discretizar.

Cuando usamos kmeans en general se mejora al aumentar o no se empeora enormemente (pues la forma de asociar a estos bins es más inteligente y menos sensible a la aleatoriedad). Podemos ver este comportamiento con naïve bayes y en menor medida con los otros dos que no necesitan de muchos más de 3 recipientes (pues buscamos clasificar solo dos estados o clusters).

Al hacer uso de los quantiles como la asociacion a los bins ya no depende tanto de un procedimiento sino reparte en partes iguales aquí vemos que aumentar no siempre aporta algo bueno (en ocasiones empeorando mucho como es con el caso de kmeans).

En lo que a test error respecta la regresión logistica se comporta muy bien en general y siempre o aumenta o pierde poco.

```python
fig, axs = plt.subplots(1, 2, figsize=(14, 6))

# Cross Validation Error Avg vs. Number of Bins para 'kmeans'
kmeans_df = sorted_df[sorted_df['Discretization Method'] == 'kmeans']
kmeans_df_sorted = kmeans_df.sort_values('Num Bins')

for model in kmeans_df_sorted['Model'].unique():
    model_data = kmeans_df_sorted[kmeans_df_sorted['Model'] == model]
    axs[0].plot(model_data['Num Bins'], model_data['Cross Validation Error Avg'], marker='o', label=model)

axs[0].set_title('Cross Validation Error Avg vs. Number of Bins (KMeans Discretization)')
axs[0].set_xlabel('Number of Bins')
axs[0].set_ylabel('Cross Validation Error Avg')
axs[0].legend()
axs[0].grid(True)

quantile_df = sorted_df[sorted_df['Discretization Method'] == 'quantile']
quantile_df_sorted = quantile_df.sort_values('Num Bins')

for model in quantile_df_sorted['Model'].unique():
    model_data = quantile_df_sorted[quantile_df_sorted['Model'] == model]
    axs[1].plot(model_data['Num Bins'], model_data['Cross Validation Error Avg'], marker='o', label=model)

axs[1].set_title('Cross Validation Error Avg vs. Number of Bins (Quantile Discretization)')
axs[1].set_xlabel('Number of Bins')
axs[1].set_ylabel('Cross Validation Error Avg')
axs[1].legend()
axs[1].grid(True)

plt.tight_layout()
plt.show()
```

La validación cruzada pone en mejor posicion a KNN sobre la regresión logística en general pues no llega a valles tan profundos y tiene una tendencia de crecimiento generalizada muy positiva al aumentar el número de recipientes de clasificación.

La regresión logística sigue de cerca en la discretización kmeans (más inteligente) y demuestra un comportamiento más erratico (con un gran valle) en el método de quantiles el cual tiene sentido por ser este método más arbitrario.

Naïve bayes excepcionalmente en los quantiles con 2 recipientes se acerca bastante a su potencial simple (sin superarlo) pero después áctua también de forma erràtica.

En nuestro caso vemos que si bien mejora algunas metricas en casos limitados como el de la regresion logistica (dejando igual su test error y mejorando el cruzado con quantile y 4 cubos) o mejorando en general como es el caso de con knn (también con quantile) no es justificable pues las mejora de forma marginal.

No somos capaces de ver con las configuraciones que hemos intentado un cambio sustancial como para justificar el cambio. Si explorasemos más posibilidades dentro de los recipientes que tomamos o métodos de clasificacion podriamos encontrar algo más interesante pero con la información actual no hay evidencia para confirmar que discretizar tenga un impacto suficiente como para considerarlo.

En general se valoraria positivamente de la discretización la capacidad de reducir la complejidad del modelo. Pero creo que este modelo es suficientemente simple y no se beneficia de esto. La reducción de valores atípicos si bien resulta positiva como estandarizamos los datos previamente puede que ya lidiemos con esta problemática en la ejecución de control del apartado B. A nivel de rendimiento computacional el dataset y la decisión es muy pequeña y por tanto tampoco se puede apreciar una gran mejoría de eficiencia. Esto sumado a la potencial perdida de información y la sensibilidad al elejir método de discretización (pueden variar mucho en eficiencia) nos hace confirmar que para este caso no es beneficioso lo que la discretización tiene que ofrecer.
